```python
books_bins = np.histogram_bin_edges(college['Books'], bins='fd')
FTU_bins = np.histogram_bin_edges(college['F.Undergrad'], bins='fd')
PTU_bins = np.histogram_bin_edges(college['P.Undergrad'], bins='fd')
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 8))
college['Books'].plot.hist(bins=19, ax=axes[0, 0], title='1: Costs of course literature, 19 bins')
college['Books'].plot.hist(bins='fd', ax=axes[1, 0], title=f'2: Costs of course literature, {len(books_bins) - 1} bins')
college['F.Undergrad'].plot.hist(bins=19, ax=axes[0, 1], title='3: Full-time undergrads, 19 bins')
college['F.Undergrad'].plot.hist(bins='fd', ax=axes[1, 1], title=f'4: Full-time undergrads, {len(FTU_bins) - 1} bins')
college['P.Undergrad'].plot.hist(bins=19, ax=axes[0, 2], title='5: Part-time undergrads, 19 bins')
college['P.Undergrad'].plot.hist(bins='fd', ax=axes[1, 2], title=f'6: Part-time undergrads, {len(PTU_bins) - 1} bins')
```

I de övre tre histogrammen har jag samlat datan i 19 bins vilket innebär att det skulle vara drygt 40 skolor per stapel om alla värden vore jämnt fördelade och i de undre tre har jag optimerat antalet bins med Freedman–Diaconis-metoden. Nu är ju inte alla värden jämnt fördelade förstås och med dessa grafer kan vi då utröna inom vilka spann de flesta skolor ligger.

1: Visar kostnader för kurslitteraturen med 19 bins<br>

4: Visar antalet fulltidsstuderande på grundutbildningar med det optimala antalet bins beräknat med Freedman–Diaconis-metoden, i detta fallet 49st
