# Kostnader för elitskolor

Efter att ha tagit fram vilka skolor som är elitskolor så blev jag intresserad av att jämföra dessa skolors studenters utgifter gentemot de andra skolornas, hädanefter kallade pöbelskolor, studenter.

```python
college.loc[college['Grad.Rate'] > 100, 'Grad.Rate'] = np.nan # Behövligt då det är ett exempel som har 118% Grad.Rate så det är felaktig data, bättre att hoppa över än att försöka gissa vad som vore korrekt
college['Cost'] = college['Outstate'] + college['Room.Board'] + college['Books'] + college['Personal'] + college['Expend']
college['Value'] = college['Cost'] / college['Grad.Rate']
```

```python
elitskolor = college[college['Elite'] == 'Yes'].copy()
pöbelskolor = college[college['Elite'] == 'No'].copy()
dyraste_elit = elitskolor.sort_values(by='Cost', ascending=False).head(1)
billigaste_elit = elitskolor.sort_values(by='Cost', ascending=True).head(1)
dyraste_pöbel = pöbelskolor.sort_values(by='Cost', ascending=False).head(1)
billigaste_pöbel = pöbelskolor.sort_values(by='Cost', ascending=True).head(1)
```

```python
print(f'Studenter på elitskolors genomsnittliga utgifter är ${elitskolor['Cost'].mean():.2f}')
print(f'Studenter på pöbelskolors genomsnittliga utgifter är ${pöbelskolor['Cost'].mean():.2f}')
print(f'Den dyraste elitskolan är {dyraste_elit.index[0]} där studenternas genomsnittliga utgifter är ${elitskolor['Cost'].max()}')
print(f'Den dyraste pöbelskolan är {dyraste_pöbel.index[0]} där studenternas genomsnittliga utgifter är ${pöbelskolor['Cost'].max()}')
print(f'Den billigaste elitskolan är {billigaste_elit.index[0]} där studenternas genomsnittliga utgifter är ${elitskolor['Cost'].min()}')
print(f'Den billigaste pöbelskolan är {billigaste_pöbel.index[0]} där studenternas genomsnittliga utgifter är ${pöbelskolor['Cost'].min()}')

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
elitskolor['Cost'].plot.hist(bins=15, ax=axes[0], title='Elitskolor')
pöbelskolor['Cost'].plot.hist(bins=15, ax=axes[1], title='Pöbelskolor')
```

Här ser vi att utgifterna för studenter på elitskolorna är nästan dubbelt så höga som de för pöbelskolorna i genomsnitt, även om min och max är relativt nära varandra. Orsaken till detta är tydlig tack vare histogrammet som klart visar att de flesta pöbelskolor ligger runt \$25,000 med snabb upp- och nedrampning samt få utstickare medans elitskolorna har en lugnare upprampning till där det största antalet skolor hittas runt \$50,000 och sedan en snabb nedrampning med några få utstickare. 

# Examensratio

Vi kan sedan jämföra examensration för elitskolor gentemot pöbelskolor för att försöka komma fram till vilken skola som är bäst till priset.

```python
bäst_elit = elitskolor.sort_values(by=['Grad.Rate', 'Cost'], ascending=[False, True]).head(1)
bäst_pöbel = pöbelskolor.sort_values(by=['Grad.Rate', 'Cost'], ascending=[False, True]).head(1)
```

```python
print(f'Genomsnittlig examensratio för elitskolor: {elitskolor['Grad.Rate'].mean():.2f}%')
print(f'Genomsnittlig examensratio för pöbelskolor: {pöbelskolor['Grad.Rate'].mean():.2f}%')
print(f'Den mest prisvärda elitskolan är {bäst_elit.index[0]} som kostar ${bäst_elit['Cost'].max()} och har en examensratio på {bäst_elit['Grad.Rate'].max()}%')
print(f'Den mest prisvärda pöbelskolan är {bäst_pöbel.index[0]} som kostar ${bäst_pöbel['Cost'].max()} och har en examensratio på {bäst_pöbel['Grad.Rate'].max()}%')

fig, axes = plt.subplots(2, 3, figsize=(18, 8))
elitskolor['Grad.Rate'].plot.hist(bins=15, ax=axes[0, 0], title='Elitskolor examensratio')
pöbelskolor['Grad.Rate'].plot.hist(bins=15, ax=axes[1, 0], title='Pöbelskolor examensratio')
elitskolor['PhD'].plot.hist(bins=15, ax=axes[0, 1], title='Elitskolor PhD')
pöbelskolor['PhD'].plot.hist(bins=15, ax=axes[1, 1], title='Pöbelskolor PhD')
elitskolor['Terminal'].plot.hist(bins=15, ax=axes[0, 2], title='Elitskolor Terminal')
pöbelskolor['Terminal'].plot.hist(bins=15, ax=axes[1, 2], title='Pöbelskolor Terminal')
```

Detta gav väldigt intressanta resultat då man klart ser att elitskolor överlag har bättre examensratio än pöbelskolor, men att både inom elitskolorna och pöbelskolorna så är det den billigaste skolan som har (delat) bäst examensratio. Trots att Missouri Southern State College på pappret ser ut att vara mycket sämre än majoriteten av elitskolorna och fortfarande sämre än den lägst rankade elitskolan när det kommer till lärarkårens PhD/Terminal-ratio samt top 10/25% så presterar den lika bra eller bättre än dem och är samtidigt den billigaste skolan av alla skolorna. Med det i åtanke så klassar jag därmed Missouri Southern State College som det mest prisvärda universitetet, det innebär ju dock inte att det är det bästa universitetet då saker som bland annat prestige spelar roll för hur stort värde en examen från en viss skola har.

```python
top50_value = college.sort_values('Value', ascending=True).head(50)
bot50_value = college.sort_values('Value', ascending=False).head(50)
elite_in_topvalue = top50_value['Elite'].value_counts()
elite_in_botvalue = bot50_value['Elite'].value_counts()
print(f'Fördelningen av elitskolor bland de 50 mest prisvärda skolorna:\n{elite_in_topvalue.to_frame().to_string(header=False)}\n')
print(f'Fördelningen av elitskolor bland de 50 minst prisvärda skolorna:\n{elite_in_botvalue.to_frame().to_string(header=False)}\n')
print(f'Elitskolors genomsnittliga prisvärde: {elitskolor['Value'].mean()}')
print(f'Pöbelskolors genomsnittliga prisvärde: {pöbelskolor['Value'].mean()}')

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
top50_value.boxplot(column='Value', by='Elite', ax=axes[0])
axes[0].set_title('Top 50 mest prisvärda')
bot50_value.boxplot(column='Value', by='Elite', ax=axes[1])
axes[1].set_title('Top 50 minst prisvärda')
college.boxplot(column='Value', by='Elite', ax=axes[2])
axes[2].set_title('Alla skolors prisvärde')
plt.suptitle('')
```

Här ser vi en tydlig korrelation mellan bäst prisvärde (ju lägre siffra desto bättre) och pöbelskolor, men inte att samma speglas i lågt prisvärde där det är jämnare fördelat även om pöbelskolor leder även där. Av detta kan man uppfatta det som att ifall målet bara är att ta examen för examens skull så är pöbelskolor överlag ett bättre alternativ men att man då behöver forska lite så att man väljer "rätt" pöbelskola då de flesta av de med sämst prisvärde är utstickare bland pöbelskolorna. Överlag är genomsnittligt prisvärde relativt nära varandra för både elitskolor och pöbelskolor.
